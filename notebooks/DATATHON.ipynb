# DF - Criando uma tabela única

```python
# Vou criar uma unica tabela para realizar a construção das analises e preparar o modelo
# Criando a coluna ANO
df_2022['ano'] = 2022
df_2023['ano'] = 2023
df_2024['ano'] = 2024

#incluindo a coluna ano_nascimento para calculo da idade
df_2022['ano_nascimento'] = pd.to_numeric(df_2022['Ano nasc'], errors='coerce')
df_2023['ano_nascimento'] = pd.to_datetime(df_2023['Data de Nasc'], errors='coerce').dt.year
df_2024['ano_nascimento'] = pd.to_datetime(df_2024['Data de Nasc'], errors='coerce').dt.year

# Mapeamento das colunas (De: Para)
# Focando apenas nas colunas necessárias para as análises e para o modelo preditivo
map_2022 = {
    'Fase': 'fase', 'INDE 22': 'inde', 'IAA': 'iaa', 'IEG': 'ieg', 'IPS': 'ips', 'IDA': 'ida', 'IPV': 'ipv', 'IAN': 'ian','Defas':'defasagem', 'ano': 'ano','Pedra 22':'pedra','Gênero': 'genero', 'Instituição de ensino': 'instituicao_ensino', 'ano_nascimento': 'ano_nascimento'
    # 2022 não tem IPP
}

map_2023 = {
    'Fase': 'fase', 'INDE 2023': 'inde', 'IAA': 'iaa', 'IEG': 'ieg', 'IPS': 'ips', 'IPP': 'ipp', 'IDA': 'ida', 'IPV': 'ipv', 'IAN': 'ian','Defasagem':'defasagem', 'ano': 'ano','Pedra 2023':'pedra','Gênero': 'genero', 'Instituição de ensino': 'instituicao_ensino', 'ano_nascimento': 'ano_nascimento'
}

map_2024 = {
    'Fase': 'fase', 'INDE 2024': 'inde', 'IAA': 'iaa', 'IEG': 'ieg', 'IPS': 'ips', 'IPP': 'ipp', 'IDA': 'ida', 'IPV': 'ipv', 'IAN': 'ian','Defasagem':'defasagem', 'ano': 'ano','Pedra 2024':'pedra','Gênero': 'genero', 'Instituição de ensino': 'instituicao_ensino', 'ano_nascimento': 'ano_nascimento'
}

# Renomeando e filtrando
df_22_clean = df_2022[list(map_2022.keys())].rename(columns=map_2022).copy()
df_23_clean = df_2023[list(map_2023.keys())].rename(columns=map_2023).copy()
df_24_clean = df_2024[list(map_2024.keys())].rename(columns=map_2024).copy()

# Empilhamento das tabelas
df_consolidado = pd.concat([df_22_clean, df_23_clean, df_24_clean], ignore_index=True)

# Cálculo exato da Idade
df_consolidado['idade'] = df_consolidado['ano'] - df_consolidado['ano_nascimento']

# 5. Forçando os tipos para numérico e lidando com erros do Excel (#DIV/0!/#N/A)
cols_numericas = ['inde', 'iaa', 'ieg', 'ips', 'ipp', 'ida', 'ipv', 'ian','defasagem','idade']

for col in cols_numericas:
    if col in df_consolidado.columns:
        # Se a coluna for do tipo texto (object), primeiro ajustamos a vírgula
        if df_consolidado[col].dtype == 'object':
            # Transforma em string para garantir que o replace funcione e troca vírgula por ponto
            df_consolidado[col] = df_consolidado[col].astype(str).str.replace(',', '.')
        
        # O errors='coerce' transforma '#DIV/0!' ou qualquer outro erro de texto em NaN (nulo)
        df_consolidado[col] = pd.to_numeric(df_consolidado[col], errors='coerce')

print(f"Total de registros: {df_consolidado.shape[0]}")
print("\nAmostra dos dados:")
display(df_consolidado.head())
```

```python
df_consolidado.info()
#mesmo após o tratamento temos linhas vazias, porém como não há dados nulos na defasagem, escolhi manter por enquanto.
```

```python
#padronizando a coluna fase
import re

def extrair_numero_fase(valor):
    if pd.isna(valor):
        return valor
    
    # Converte para string e remove espaços
    texto = str(valor).strip().upper()
    
    # Trata o caso "ALFA" como fase 0
    if 'ALFA' in texto:
        return 0
    
    # Extrai apenas os números (ex: "Fase 1" -> 1, "1 - Fase 1" -> 1)
    numeros = re.findall(r'\d+', texto)
    if numeros:
        return int(numeros[0])
    
    return np.nan


df_consolidado['fase'] = df_consolidado['fase'].apply(extrair_numero_fase)

df_consolidado['fase'] = pd.to_numeric(df_consolidado['fase'], errors='coerce')

print(df_consolidado['fase'].unique())
```

```python
#o ano de 2022 não trouxe o ipp, tento popular o valor realizando o calculo reverso 
# Cria um filtro para identificar exatamente quais linhas precisam de cálculo
# Condições: Ano 2022, Fase diferente de 8, e onde o IPP ainda está nulo
filtro_calculo = (
    (df_consolidado['ano'] == 2022) & 
    (df_consolidado['fase'] != 8) & 
    (df_consolidado['ipp'].isna())
)

# Calcula a soma ponderada dos outros indicadores conhecidos
soma_ponderada_outros = (
    (df_consolidado['ian'] * 0.1) + 
    (df_consolidado['ida'] * 0.2) + 
    (df_consolidado['ieg'] * 0.2) + 
    (df_consolidado['iaa'] * 0.1) + 
    (df_consolidado['ips'] * 0.1) + 
    (df_consolidado['ipv'] * 0.2)
)

# Aplica a fórmula reversa apenas nas linhas que passaram no filtro
df_consolidado.loc[filtro_calculo, 'ipp'] = (df_consolidado.loc[filtro_calculo, 'inde'] - soma_ponderada_outros.loc[filtro_calculo]) / 0.1

# Arredondamento e formatação
# Arredonda para 1 casa decimal para manter o padrão das outras notas
df_consolidado['ipp'] = df_consolidado['ipp'].round(1)

display(df_consolidado[(df_consolidado['ano'] == 2022) & (df_consolidado['fase'] != 8)][['ano', 'fase', 'inde', 'ipp']].head())
```

```python
#criando a coluna categorica para mostrar defasagem conforme regra de negocio
mapeamento_ian = {
    10.0: 'Em fase',
    5.0: 'Moderada',
    2.5: 'Severa'
}

df_consolidado['perfil_defasagem'] = df_consolidado['ian'].map(mapeamento_ian)

df_consolidado.head(10)
```

# Visualização dos dados

```python
# 1. Adequação do nível (IAN)
cores_defasagem = {'Em fase': '#2ca02c', 'Moderada': '#ff7f0e', 'Severa': '#d62728'}
ordem_defasagem = ['Em fase', 'Moderada', 'Severa']

plt.figure(figsize=(10, 6))

ax = sns.countplot(
    data=df_consolidado, 
    x='ano', 
    hue='perfil_defasagem', 
    hue_order=ordem_defasagem,
    palette=cores_defasagem
)

plt.title('Evolução do Perfil de Defasagem dos Alunos (IAN)', fontweight='bold', pad=15)
plt.xlabel('Ano')
plt.ylabel('Quantidade de Alunos')
plt.legend(title='Nível de Defasagem', bbox_to_anchor=(1.05, 1), loc='upper left')

for p in ax.patches:
    altura = p.get_height()
    if not np.isnan(altura) and altura > 0:
        ax.annotate(f'{int(altura)}', 
                    (p.get_x() + p.get_width() / 2., altura), 
                    ha='center', va='bottom', 
                    fontsize=11, color='black', xytext=(0, 3), 
                    textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
# 2. Desempenho acadêmico (IDA) - Visão Dividida

# Gráfico 1: IDA ao longo dos Anos
plt.figure(figsize=(8, 5))
ax1 = sns.barplot(data=df_consolidado, x='ano', y='ida', color='#4C72B0', errorbar=None)

plt.title('2.1 Evolução do Desempenho Acadêmico Médio (IDA) por Ano', fontweight='bold', pad=15)
plt.xlabel('Ano')
plt.ylabel('Média da Nota IDA')
plt.ylim(0, 10) 

for p in ax1.patches:
    ax1.annotate(f'{p.get_height():.1f}', 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha='center', va='bottom', fontsize=12, fontweight='bold',
                 xytext=(0, 5), textcoords='offset points')

plt.tight_layout()
plt.show()

# Gráfico 2.2: IDA ao longo das Fases e Anos (Linhas)
plt.figure(figsize=(12, 6))

ax2 = sns.lineplot(
    data=df_consolidado, 
    x='fase', 
    y='ida', 
    hue='ano', 
    palette=['#1f77b4', '#ff7f0e', '#2ca02c'], # Azul (2022), Laranja (2023), Verde (2024)
    marker='o', 
    markersize=8,
    linewidth=2.5,
    errorbar=None
)

plt.title('2.2 Evolução do Desempenho Acadêmico Médio (IDA) por Fase e Ano', fontweight='bold', pad=15)
plt.xlabel('Fase')
plt.ylabel('Média da Nota IDA')
plt.ylim(0, 10)

fases_unicas = sorted(df_consolidado['fase'].dropna().unique())
plt.xticks(fases_unicas)

plt.legend(title='Ano', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
#3. Engajamento nas atividades (IEG)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico 1: IEG x IDA
sns.regplot(data=df_consolidado, x='ieg', y='ida', ax=axes[0],
            scatter_kws={'alpha':0.4, 'color':'#1f77b4'}, 
            line_kws={'color':'red', 'linewidth':3})
axes[0].set_title('Engajamento vs Desempenho (IDA)', fontweight='bold')
axes[0].set_xlabel('Nota de Engajamento (IEG)')
axes[0].set_ylabel('Nota de Desempenho (IDA)')
axes[0].set_ylim(0, 11)

# Gráfico 2: IEG x IPV
sns.regplot(data=df_consolidado, x='ieg', y='ipv', ax=axes[1],
            scatter_kws={'alpha':0.4, 'color':'#2ca02c'}, 
            line_kws={'color':'red', 'linewidth':3})
axes[1].set_title('Engajamento vs Ponto de Virada (IPV)', fontweight='bold')
axes[1].set_xlabel('Nota de Engajamento (IEG)')
axes[1].set_ylabel('Nota Ponto de Virada (IPV)')
axes[1].set_ylim(0, 11)

plt.tight_layout()
plt.show()
```

```python
#4. Autoavaliação (IAA)
plt.figure(figsize=(8, 6))
correlacao_iaa = df_consolidado[['iaa', 'ida', 'ieg']].corr()
sns.heatmap(correlacao_iaa, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('Correlação: Autoavaliação x Desempenho x Engajamento')
plt.show()
```

```python
#5. Aspectos psicossociais (IPS)
plt.figure(figsize=(12, 6))
df_tendencia = df_consolidado.groupby('ano')[['ips', 'ida', 'ieg']].mean().reset_index()
df_melted = df_tendencia.melt(id_vars='ano', var_name='Indicador', value_name='Média')

sns.lineplot(data=df_melted, x='ano', y='Média', hue='Indicador', marker='s', linewidth=2.5)
plt.title('Tendência Média Temporal: Psicossocial (IPS) vs Acadêmico (IDA) e Engajamento (IEG)')
plt.ylabel('Média das Notas')
plt.xlabel('Ano')
plt.show()
```

```python
# 6. Aspectos psicopedagógicos (IPP)

plt.figure(figsize=(10, 6))

ordem_defasagem = ['Em fase', 'Moderada', 'Severa']
cores_defasagem = {'Em fase': '#2ca02c', 'Moderada': '#ff7f0e', 'Severa': '#d62728'}

ax6 = sns.boxplot(
    data=df_consolidado, 
    x='perfil_defasagem', 
    y='ipp', 
    order=ordem_defasagem, 
    palette=cores_defasagem,
    width=0.5,
    fliersize=5 
)

plt.title('6. Avaliação Psicopedagógica (IPP) por Nível de Defasagem (IAN)', fontweight='bold', pad=15)
plt.xlabel('Perfil de Defasagem (IAN)')
plt.ylabel('Nota Psicopedagógica (IPP)')
plt.ylim(0, 11)

plt.tight_layout()
plt.show()
```

```python
#6. Aspectos psicopedagógicos (IPP)
plt.figure(figsize=(10, 6))
sns.regplot(data=df_consolidado, x='ipp', y='ian', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Confirmação da Defasagem: IPP x IAN')
plt.xlabel('Aspectos Psicopedagógicos (IPP)')
plt.ylabel('Adequação de Nível (IAN)')
plt.show()
```

```python
#7. Ponto de virada (IPV)
plt.figure(figsize=(10, 5))
corr_ipv = df_consolidado[['ida', 'ieg', 'ips', 'ipp', 'iaa', 'ian', 'ipv']].corr()['ipv'].drop('ipv').sort_values(ascending=False)

sns.barplot(x=corr_ipv.values, y=corr_ipv.index, palette='magma')
plt.title('Quais indicadores mais influenciam o Ponto de Virada (IPV)?')
plt.xlabel('Coeficiente de Correlação')
plt.ylabel('Indicadores')
plt.show()
```
